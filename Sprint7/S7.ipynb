```python
#BMI Calculator

def BMI_calculate(a, b):
    return a/b**2

def BMI_sort(z):
    if z < 18.5:
        return 'underweight'
    elif z < 25:
        return 'normal weight'
    elif z < 30:
        return 'overweight'
    else:
        return 'obese'
    
def main():
    w = float(input('Please intorduce your weight in kgs:'))
    h_cm = float(input('Thank you, please now introduce your height in cms:'))
    h_m = h_cm/100
    
    bmi = BMI_calculate(w, h_m)
    clas = BMI_sort(bmi)

    print('Your BMI is {:.2f}, and you are'.format(bmi), clas)

main()
```

```python
import string

def split(t):
    taslist = t.split()
    return taslist

def trans(t):
    transalte1 = str.maketrans('','', string.punctuation)
    transalte2 = str.maketrans('','','¡')
    transalte3 = str.maketrans('','','¿')

    t1 = t.translate(transalte1)
    t2 = t1.translate(transalte2)
    t3 = t2.translate(transalte3)

    return t3

def counter():
    t = input('Please input the text you want to count:').lower()

    t = trans(t)
    t = split(t)

    word_counter = {}

    for word in t:
        word_counter[word] = t.count(word)
    
    return word_counter

counter()
```

```python
import re

def convert_mixed_fraction(string):
    # Using regular expression to match mixed fractions
    match = re.match(r'^(\d+)-(\d+)/(\d+)$', string)
    if match:
        whole, numerator, denominator = match.groups()
        return float(whole) + float(numerator) / float(denominator)
    else:
        return string

client_list = ['1.3', 'one', '1e10', 'seven', '3-1/2', ('2', 1, 1.4, 'not-a-number'), [1, 2, '3', '3.4']]

floats = []
other = []

for a in client_list:
    if isinstance(a, (list, tuple)):
        for b in a:
            if isinstance(b, str):
                b = convert_mixed_fraction(b)
            try:
                float_b = float(b)
                floats.append(float_b)
            except ValueError:
                other.append(b)
    else:
        if isinstance(a, str):
            a = convert_mixed_fraction(a)
        try:
            float_a = float(a)
            floats.append(float_a)
        except ValueError:
            other.append(a)

print("Floats:", floats)
print("Other:", other)
```

```python
import string

def split(t):
    taslist = t.split()
    return taslist

def trans(t):
    transalte1 = str.maketrans('','', string.punctuation)
    transalte2 = str.maketrans('','','¡')
    transalte3 = str.maketrans('','','¿')

    t1 = t.translate(transalte1)
    t2 = t1.translate(transalte2)
    t3 = t2.translate(transalte3)

    return t3

def counter():
    t = input('Please input the text you want to count:').lower()

    t = trans(t)
    t = split(t)

    orded_dict = {}

    for word in t:
        fl = word[0]
        if fl not in orded_dict:
            orded_dict[fl] = {}
        if word not in orded_dict[fl]:
            orded_dict[fl][word] = t.count(word)
            orded_dict[fl] = dict(sorted(orded_dict[fl].items()))

    orded_dict = dict(sorted(orded_dict.items()))

    return orded_dict


counter()
```
