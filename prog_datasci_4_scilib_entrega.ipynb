$$Cost_{final} = Cost_{energia} + Cost_{potència} - Cost_{energiaexcedentària}$$

Utilitza la tarifa 2 i tingues en compte el preu dels excedents de l'apartat anterior (0.08 €/KWh). **0.5 punts**

````python
#Resposta
import calendar

monthly_grouped_tp = df.groupby(['month', 'time_period']).agg({'exported_energy_kwh': 'sum', 'purchased_energy_kwh': 'sum'}).reset_index() #Creem una agrupació mensual i de time period de la suma de les variables desitjades

tarifa = {1: 0.18, 2: 0.12, 3: 0.08} #Mapejem la tarifa

monthly_grouped_tp['cost_per_kwh'] = monthly_grouped_tp['time_period'].map(tarifa) #Afegim columna on introduim els valors de les tarifes

monthly_grouped_tp['cost'] = monthly_grouped_tp['purchased_energy_kwh']*monthly_grouped_tp['cost_per_kwh'] #Calculem el cost de comprar energia en cada franja i creem columna amb aquestes dades

months = {1: 31, 2: 28, 3: 31, 4: 30, 5: 31, 6: 30, 7: 31, 8: 31, 9: 30, 10: 31, 11: 30, 12: 31} #Mapejem els dies de cada mes

monthly_grouped_tp['days'] = monthly_grouped_tp['month'].map(months) #Intorduim columna amb nombre de dies per mes utilitzant el diccionari creat

monthly_grouped_tp['balance'] = 2*0.11*monthly_grouped_tp['days']+monthly_grouped_tp['cost']-monthly_grouped_tp['exported_energy_kwh']*0.08 #Calculem el balanç final en una columna

monthly_price_t2 = monthly_grouped_tp.groupby('month')['balance'].sum() #Sumem el resultats de la columna i els agrupem epr mes

print(monthly_price_t2) #Mostrem resultat
````

- $\color{orange}{\text{EG}}$Mostra les estadístiques (el punt d'intersecció, el coeficient de correlació de Pearson i el p-value) que et torna la funció. Podries explicar quin significat té el valor del pendent? I el valor de $R^2$?Hi ha un bon ajust entre el dia de l'any i l'energia produida? Hi ha una tendència negativa o positiva? Raona les respostes. (**0.5 punts**)


````python
#Resposta
from scipy.stats import linregress

df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d') #Assegurem format

jan1_jun23_data = df.loc[(df['date'] >= '2023-01-01') & (df['date'] <= '2023-06-23')].sort_values('date') #Filtrem y ordenem

jan1_jun23_data_2 = jan1_jun23_data.groupby('date')['generated_energy_kwh'].sum()

x_value_lr = range(len(jan1_jun23_data_2)) #Definim x i y per utilitzar lineregress
y_value_lr = jan1_jun23_data_2.values

slope, intercept, r_value, p_value, std_err = linregress(x_value_lr, y_value_lr) #Utilitzem lineregress

print("Slope:", slope)
print("Intercept:", intercept)
print("R-squared:", r_value**2)
print("P-value:", p_value)
print("Standard error:", std_err)

#Slope = Mostra el canvi mig de la variable y (generated) en funcio del canvi unitari de la variable x (dies). Aixó ens indica cuan varia la generació dia rere dia. En aquest cas 0.045 de forma postiva, es a dir que augmenta
#Intercept = El valor de y cuan x = 0. En aquest cas el valor de y = 2.6kwh el dia 1 de gener.
#r**2 = Indica la proporcio de la variació de y que es deguda a la relació entre x i y. Aquest valor va de 0 a 1, com mes s'apropa a 1 mes relació esixteix. En el nostre cas un 66% de la variació de y es deguada a x
#p_value = Indica si la dependencia entre x i y es forta, es busquen valors inferiors a 0.05, en el nostre cas el p_value es molt petit així que existeix una forta relació.
#Standard error = Un standard error baix indica una alta precisó dels coeficients calculats, en aquest cas es baix així que es positiu.
````
