```python
import mysql.connector
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Connect to the MySQL database
conn = mysql.connector.connect(host='localhost',
                                         database='world',
                                         user='root',
                                         password='root'
)

query = "SELECT COUNT(ROUND(lifeexpectancy, 0)), ROUND(lifeexpectancy, 0) FROM country GROUP BY ROUND(lifeexpectancy, 0)"
cursor = conn.cursor()
cursor.execute(query)

data = cursor.fetchall()
columns = [i[0] for i in cursor.description]
df = pd.DataFrame(data, columns=columns)

sns.histplot(data=df, x='ROUND(lifeexpectancy, 0)')
plt.title("Count of life expectancy")
plt.xlabel("Life Expectancy")
plt.show()

cursor.close()
conn.close()
```

```python
import mysql.connector
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(rc={'figure.figsize':(20, 20)})

# Step 1: Connect to the MySQL database
conn = mysql.connector.connect(host='localhost',
                                         database='world',
                                         user='root',
                                         password='root'
)

query = "SELECT surfacearea, population FROM country"
cursor = conn.cursor()
cursor.execute(query)

data = cursor.fetchall()
columns = [i[0] for i in cursor.description]
df = pd.DataFrame(data, columns=columns)

sns.scatterplot(data=df, x='surfacearea', y='population')
plt.title("SurfaceArea vs Population")
plt.xlabel("surfacearea")
plt.ylabel("population")
plt.show()

cursor.close()
conn.close()
```

```python
import mysql.connector
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Connect to the MySQL database
conn = mysql.connector.connect(host='localhost',
                                         database='world',
                                         user='root',
                                         password='root'
)

query = "SELECT SUM(ROUND(poplanguage, 2)) AS totalspeakers, language FROM (SELECT countrylanguage.*, population, ((percentage/100)*population) AS poplanguage FROM country, countrylanguage WHERE country.code = countrylanguage.countrycode) AS t1 GROUP BY language"
cursor = conn.cursor()
cursor.execute(query)

data = cursor.fetchall()
columns = [i[0] for i in cursor.description]
df = pd.DataFrame(data, columns=columns)

sns.barplot(data=df, x='language', y='COUNT(language)')
plt.title("Languages World")
plt.xlabel("Language")
plt.ylabel("Speaking countries")
plt.show()

cursor.close()
conn.close()
```
