```python
import mysql.connector
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Connect to MySQL
conn = mysql.connector.connect(host='localhost',
                               database='iris_dataset',
                               user='root',
                               password='Peanuts@1991')

# Function to fetch data from MySQL and convert it into a DataFrame
def fetch_data_to_dataframe(connection, query):
    cursor = connection.cursor()
    cursor.execute(query)
    data = cursor.fetchall()
    columns = [i[0] for i in cursor.description]
    df = pd.DataFrame(data, columns=columns)
    return df

# Query to select data from a table (replace 'your_table' with the actual table name)
query = "SELECT * FROM iris"

# Fetch data from MySQL and convert it into a DataFrame
df = fetch_data_to_dataframe(conn, query)

# Close the connection
conn.close()

# Now you have your data in a Pandas DataFrame
sns.violinplot(data=df, x='class', y='sepal_length', inner=None)
```

```python
import mysql.connector
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Connect to MySQL
conn = mysql.connector.connect(host='localhost',
                               database='bussines',
                               user='root',
                               password='Peanuts@1991')

# Function to fetch data from MySQL and convert it into a DataFrame
def fetch_data_to_dataframe(connection, query):
    cursor = connection.cursor()
    cursor.execute(query)
    data = cursor.fetchall()
    columns = [i[0] for i in cursor.description]
    df = pd.DataFrame(data, columns=columns)
    return df

# Query to select data from a table (replace 'your_table' with the actual table name)
query = "SELECT * FROM company, user, product"

# Fetch data from MySQL and convert it into a DataFrame
df = fetch_data_to_dataframe(conn, query)

# Close the connection
conn.close()

# Now you have your data in a Pandas DataFrame
print(df.head())


sns.barplot(x='country')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame containing the 'country' column

# Count occurrences of each country
country_counts = df['country'].value_counts()

# Create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=country_counts.index, y=country_counts.values)
plt.title('Number of Records per Country')
plt.xlabel('Country')
plt.ylabel('Number of Records')
plt.xticks(rotation=45)  # Rotate x-labels for better readability
plt.tight_layout()
plt.show()
```
